```python
import datetime
import pandas as pd
import requests
```

```python
todaysDate = datetime.datetime.now()
delta = datetime.timedelta(days=30)
newdate = todaysDate - delta
key = newdate.strftime("%Y-%b")
```

```python
url = "https://api.msrc.microsoft.com/cvrf/v2.0/document/" + key
headers = {'Accept': 'application/json'}
response = requests.get(url, headers=headers)
json_data = response.json()
```

```python
json_data.keys()
```

```python
vulns = json_data.get('Vulnerability')
df_vulns = pd.json_normalize(vulns)
#revHist = df_vulns.get('RevisionHistory')
#df_revHist = pd.json_normalize(revHist)
#df_vulns
revision_data = pd.json_normalize(data=vulns, record_path=['RevisionHistory'])
```

```python
df_vulns = df_vulns.join(revision_data)
```
